Com a ajuda da biblioteca Sweetviz, eu quero descobrir qual a porcentagem de respostas com 7 = “Se recusou a responder” e 9 = “Não sabe” nas respostas do banco DEMO_PHQ.csv (SWEETVIZ_REPORT1.html na raiz do projeto).

```python
import sweetviz
```

```python
analise=sweetviz.analyze(df1)
analise.show_html()
```

Como a quantidade máxima de respostas com 7 e 9 não ultrapassam 1% das respostas (chega a 9 respostas com a opção 9 em DPQ010 e DPQ080), vou substituir o valor de 7 e 9 por 0 (zero), para não enviesar os valores calculados para phq9

```python
df1['DPQ010'].mask((df1['DPQ010'] == 7) | (df1['DPQ010'] == 9), 0, inplace=True)
df1['DPQ020'].mask((df1['DPQ020'] == 7) | (df1['DPQ020'] == 9), 0, inplace=True)
df1['DPQ030'].mask((df1['DPQ030'] == 7) | (df1['DPQ030'] == 9), 0, inplace=True)
df1['DPQ040'].mask((df1['DPQ040'] == 7) | (df1['DPQ040'] == 9), 0, inplace=True)
df1['DPQ050'].mask((df1['DPQ050'] == 7) | (df1['DPQ050'] == 9), 0, inplace=True)
df1['DPQ060'].mask((df1['DPQ060'] == 7) | (df1['DPQ060'] == 9), 0, inplace=True)
df1['DPQ070'].mask((df1['DPQ070'] == 7) | (df1['DPQ070'] == 9), 0, inplace=True)
df1['DPQ080'].mask((df1['DPQ080'] == 7) | (df1['DPQ080'] == 9), 0, inplace=True)
df1['DPQ090'].mask((df1['DPQ090'] == 7) | (df1['DPQ090'] == 9), 0, inplace=True)
df1.sample(3)
```

### Combinando os dataframes utilizando SEQN como chave única

```python
df3 = pd.merge(df1, df2, how = 'inner',  on = 'SEQN')
df3.shape
```

### Criando a varável phq9

```python
df3['phq9']=df3.iloc[:,1:9].sum(axis=1)
df3.sample(10)
```

### Criando a varável phq_grp

Função que atribui um grupo para o respondente

```python
def grupo_sintomas(df):
  ''' Atribui grupo de sintomas de depressão '''
  if df['phq9'] < 5:
    return 0
  elif 5 <= df['phq9'] < 10:
    return 1
  elif 10 <= df['phq9'] < 15:
    return 2
  elif 15 <= df['phq9'] < 19:
    return 3
  else:
    return 4
```

Aplicando a função para cada linha

```python
df3['phq_grp'] = df3.apply(grupo_sintomas, axis=1)
df3.sample(10)
```
