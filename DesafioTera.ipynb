# **Tratamento de Dados**

### Importando bibliotecas

```python
import pandas as pd
```

### Lendo os arquivos csv

```python
df1 = pd.read_csv('./bd/DEMO_PHQ.csv')
df2 = pd.read_csv('./bd/PAG_HEI.csv')
```

```python
df1.head(5)
```

```python
df2.head(10)
```

### Combinando os dataframes utilizando SEQN como chave única

```python
df3 = pd.merge(df1, df2, how = 'inner',  on = 'SEQN')
df3.sample(10)
```

### Criando a varável phq9

```python
df3['phq9']=df3.iloc[:,1:9].sum(axis=1)
df3.sample(10)
```

### Criando a varável phq_grp

Função que atribui um grupo para o respondente

```python
def grupo_sintomas(df):
  ''' Atribui grupo de sintomas de depressão '''
  if df['phq9'] < 5:
    return 0
  elif 5 <= df['phq9'] < 10:
    return 1
  elif 10 <= df['phq9'] < 15:
    return 2
  elif 15 <= df['phq9'] < 19:
    return 3
  else:
    return 4
```

Aplicando a função para cada linha

```python
df3['phq_grp'] = df3.apply(grupo_sintomas, axis=1)
df3.sample(10)
```
